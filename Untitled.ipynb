```python
import cv2
import os

face_cascade = cv2.CascadeClassifier(
    cv2.data.haarcascades + "haarcascade_frontalface_default.xml"
)

name = "ashvini"
path = "dataset/" + name

cap = cv2.VideoCapture(0)

count = 0

while True:
    ret, frame = cap.read()

    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    faces = face_cascade.detectMultiScale(gray, 1.1, 5)

    for (x, y, w, h) in faces:
        count += 1

        face = gray[y:y+h, x:x+w]

        cv2.imwrite(f"{path}/{count}.jpg", face)

        cv2.rectangle(frame, (x, y), (x+w, y+h), (0, 255, 0), 2)

    cv2.imshow("Capturing Faces", frame)

    if count >= 20:
        break

    if cv2.waitKey(1) == 27:
        break

cap.release()
cv2.destroyAllWindows()

print("Saved", count, "images")
```

```python
import cv2
import os
import numpy as np

path = "dataset/ashvini"

faces = []
ids = []

for image_name in os.listdir(path):

    img_path = os.path.join(path, image_name)

    img = cv2.imread(img_path, 0)

    faces.append(img)

    ids.append(1)

recognizer = cv2.face.LBPHFaceRecognizer_create()

recognizer.train(faces, np.array(ids))

recognizer.save("trainer.yml")

print("Model trained successfully")
```

```python

import cv2

recognizer = cv2.face.LBPHFaceRecognizer_create()
recognizer.read("trainer.yml")

face_cascade = cv2.CascadeClassifier(
    cv2.data.haarcascades + "haarcascade_frontalface_default.xml"
)

cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()

    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    faces = face_cascade.detectMultiScale(gray, 1.1, 5)

    for (x, y, w, h) in faces:

        id, confidence = recognizer.predict(gray[y:y+h, x:x+w])

        if id == 1:
            name = "Ashvini"
        else:
            name = "Unknown"

        cv2.rectangle(frame, (x, y), (x+w, y+h), (0, 255, 0), 2)

        cv2.putText(
            frame,
            name,
            (x, y-10),
            cv2.FONT_HERSHEY_SIMPLEX,
            1,
            (0, 255, 0),
            2
        )

    cv2.imshow("Face Recognition", frame)

    if cv2.waitKey(1) == 27:
        break

cap.release()
cv2.destroyAllWindows()
```

```python
import pandas as pd
from datetime import datetime
import os

file = "attendance.xlsx"

def mark_attendance(name):
    now = datetime.now()
    date = now.strftime("%Y-%m-%d")
    time = now.strftime("%H:%M:%S")

    new_data = pd.DataFrame([[name, date, time]],
                            columns=["Name", "Date", "Time"])

    if os.path.exists(file):
        old = pd.read_excel(file)
        updated = pd.concat([old, new_data], ignore_index=True)
        updated.to_excel(file, index=False)
    else:
        new_data.to_excel(file, index=False)

    print("Attendance saved for:", name)
```
